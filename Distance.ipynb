```python
import numpy as np
import pandas as pd
```

## 1. (X,Y) koordináták generálása

```python
a = []
for i,j in zip(range (1,100), reversed(range(1,100))):
    a.append((i,j))
a = tuple(a)
```

## 2. df-be vágom

```python
df = pd.DataFrame(a, columns=['X', 'Y'])
df = df.loc[(df["X"] <= 50)]
```

## 3. Origótól vett távolságot számítok

```python
df["Euclides"]=df["X"]**2+df["Y"]**2
df["Mark"]=df["X"]*df["Y"]
df.reset_index()
```

## 4. Ha sorba rendezem a pontjaimat az euklideszi távolság alapján (X^2 + Y^2), az ugyanazt a sorrendet adja vissza, mintha X*Y-t számolnék és azt fordított sorrendbe rendezném

```python
df["csekk"] = np.where(
    df.sort_values(by="Euclides", axis=0, ascending=True).index.values.astype(int)
    == df.sort_values(by="Mark", axis=0, ascending=False).index.values.astype(int),
    True,
    False,
)
```
